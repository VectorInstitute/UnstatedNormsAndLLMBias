```python
import os
import re
from pathlib import Path

import pandas as pd

REPO_ABS_PATH = Path(os.path.abspath("")).parent.parent.parent

DATASET_PATH = (
    f"{REPO_ABS_PATH}/unstated_norms_llm_bias/prompt_based_classification/resources/2018-Valence-oc-En-train.txt"
)
SAMPLE_WRITE_PATH = (
    f"{REPO_ABS_PATH}/unstated_norms_llm_bias/prompt_based_classification/resources/processed_semeval.tsv"
)

N_SAMPLES = 500

semeval_df = pd.read_csv(DATASET_PATH, sep="\t")
sampled_df = semeval_df.sample(N_SAMPLES)

tweets = sampled_df["Tweet"].tolist()
valences = sampled_df["Intensity Class"].tolist()
valence_map = {
    "0": "Neutral",
    "-1": "Neutral",
    "-2": "Negative",
    "-3": "Negative",
    "1": "Neutral",
    "2": "Positive",
    "3": "Positive",
}

with open(SAMPLE_WRITE_PATH, "w") as f:
    f.write("Text\tValence\n")
    for tweet, valence in zip(tweets, valences):
        mapped_valence = valence.strip()
        mapped_valence = re.sub(r":.*", "", mapped_valence)
        mapped_valence = valence_map[mapped_valence]
        f.write(f"{tweet}\t{mapped_valence}\n")
```
